```python
reviews = pd.read_csv("data/reviews.csv")
reviews
```

```python
reviews.sort_values("product_content_id").head(20)
```

```python
review_rates_mean = dict(reviews.groupby(["product_content_id", "supplier_id"])["rate"].mean())
review_rates_min = dict(reviews.groupby(["product_content_id", "supplier_id"])["rate"].min())
review_rates_max = dict(reviews.groupby(["product_content_id", "supplier_id"])["rate"].max())
review_rates_median = dict(reviews.groupby(["product_content_id", "supplier_id"])["rate"].median())

df["review_rates_mean"] = df.apply(lambda x: review_rates_mean.get((x.product_content_id, x.supplier_id), 0), axis=1)
df["review_rates_min"] = df.apply(lambda x: review_rates_min.get((x.product_content_id, x.supplier_id), 0), axis=1)
df["review_rates_max"] = df.apply(lambda x: review_rates_max.get((x.product_content_id, x.supplier_id), 0), axis=1)
#df["review_rates_median"] = df.apply(lambda x: review_rates_median.get((x.product_content_id, x.supplier_id), 0), axis=1)
```

```python
products = pd.read_csv("data/products.csv")
products
```

```python
products.columns
```

```python
products = products.drop(["product_id", "product_name", "gender_id", "gender_name", "color_id", "color_name",
               "supplier_color_name", "attributet_name", "attribute_value"], axis=1)
products
```

```python
df = df.merge(products, on=["product_content_id", "product_variant_id"])
df
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])
df.sort_values("order_date", inplace=True)
```

```python
df.is_returned.value_counts()
```

```python
train = df[df.order_date <= "2021-07-31"]
test = df[df.order_date > "2021-07-31"]
```

```python
train.to_csv("train_all.csv", index=False)
test.to_csv("test_all.csv", index=False)
```
