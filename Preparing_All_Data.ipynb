```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("transactions.csv")
df.head()
```

```python
supplier_return = pd.read_csv("supplier_return.csv")
supplier_return
```

```python
df = df.merge(supplier_return, on="supplier_id")
df
```

```python
supplier_disputed_return = pd.read_csv("supplier_disputed_return.csv")
supplier_disputed_return
```

```python
df = df.merge(supplier_disputed_return, on="supplier_id")
df
```

```python
supplier_defective_return = pd.read_csv("supplier_defective_return.csv")
supplier_defective_return
```

```python
supplier_defective_return.columns = ["supplier_id", "defective_returnRate"]
```

```python
df = df.merge(supplier_defective_return, on="supplier_id")
df
```

```python
reviews = pd.read_csv("reviews.csv")
reviews
```

```python
review_rates_mean = dict(reviews.groupby(["product_content_id", "supplier_id"])["rate"].mean())
#review_rates_mean
```

```python
review_rates_min = dict(reviews.groupby(["product_content_id", "supplier_id"])["rate"].min())
#review_rates_min
```

```python
review_rates_max = dict(reviews.groupby(["product_content_id", "supplier_id"])["rate"].max())
#review_rates_max
```

```python
review_rates_median = dict(reviews.groupby(["product_content_id", "supplier_id"])["rate"].median())
#review_rates_median
```

```python
df["review_rates_mean"] = df.apply(lambda x: review_rates_mean.get((x.product_content_id, x.supplier_id), 0), axis=1)
```

```python
df["review_rates_min"] = df.apply(lambda x: review_rates_min.get((x.product_content_id, x.supplier_id), 0), axis=1)
```

```python
df["review_rates_max"] = df.apply(lambda x: review_rates_max.get((x.product_content_id, x.supplier_id), 0), axis=1)
```

```python
df["review_rates_median"] = df.apply(lambda x: review_rates_median.get((x.product_content_id, x.supplier_id), 0), axis=1)
```
